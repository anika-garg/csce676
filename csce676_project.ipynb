# D. Exploratory Data Analysis

**Decisions**


Dataset variant: I use MIND-small instead of MIND-large

- Why: This checkpoint focuses on EDA + determining dataset feasibility. MIND-small is computationally manageable on my laptop while preserving the same schema and long-tail behavior as MIND-large.


Which split: I analyze the train split first

- Why: Train has more interaction signal than dev and is the common starting point for building baselines later.


Interaction signal: I treat history (past clicks) as the main behavioral signal for popularity/sparsity EDA

- Why: History contains confirmed clicks and is less ambiguous than impression logs (which require parsing click labels).


Missing history: I interpret missing history as “no prior history recorded” (cold-start / new user)

- Why: The dataset represents logged browsing sessions, so missing history is a realistic condition and is informative for modeling.

Cleaning strategy: I avoid aggressive text normalization beyond basic strip()

- Why: I'm trying to explore the data at this stage in the project. Heavy preprocessing can hide real-world noise that matters for bias/ethics and later modeling.

## Load data

Schema (from MIND documentation)

news.tsv: news_id, category, subcategory, title, abstract, url, title_entities, abstract_entities

behaviors.tsv: impression_id, user_id, time, history, impressions

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from collections import Counter

# File paths
NEWS_PATH = "news.tsv"
BEHAVIORS_PATH = "behaviors.tsv"
ENTITY_PATH = "entity_embedding.vec"
RELATION_PATH = "relation_embedding.vec"


news_cols = ["news_id", "category", "subcategory", "title", "abstract",
             "url", "title_entities", "abstract_entities"]

news = pd.read_csv(NEWS_PATH, sep="\t", header=None, names=news_cols)

beh_cols = ["impression_id", "user_id", "time", "history", "impressions"]
beh = pd.read_csv(BEHAVIORS_PATH, sep="\t", header=None, names=beh_cols)

print("News shape:", news.shape)
print("Behaviors shape:", beh.shape)

news.head()
```

## Data basics

Checking:

- head/tail samples
- missingness
- duplicates
- basic cardinalities (users, articles, interactions)

```python
def summarize_missing(df: pd.DataFrame) -> pd.DataFrame:
    """Return missing counts and rates for each column."""
    miss = df.isna().sum()
    rate = miss / len(df)
    out = pd.DataFrame({"missing": miss, "missing_rate": rate}).sort_values("missing_rate", ascending=False)
    return out

print("Missingness – news")
display(summarize_missing(news))

print("\nMissingness – behaviors")
display(summarize_missing(beh))

# duplicates on raw/hashable columns only
print("\nDuplicate rows – news:", news.duplicated(subset=news_cols).sum())
print("Duplicate rows – behaviors:", beh.duplicated(subset=beh_cols).sum())

print("\nUnique news_id:", news["news_id"].nunique())
print("Unique user_id:", beh["user_id"].nunique())
```

## Data cleaning and parsing

```python
beh = beh.copy()

# Parse time
# Decision: coerce errors to NaT so we can quantify bad timestamps rather than crash.
beh["time"] = pd.to_datetime(beh["time"], errors="coerce")

# Clean history
beh["history"] = beh["history"].fillna("").astype(str).str.strip()
beh["history_list"] = beh["history"].apply(lambda s: s.split() if s else [])
beh["history_len"] = beh["history_list"].apply(len)

def parse_impressions(s: str) -> list[tuple[str, int]]:
    """Parse impressions string like 'N123-1 N234-0' into list of (news_id, label)."""
    s = "" if pd.isna(s) else str(s).strip()
    if not s:
        return []
    out = []
    for token in s.split():
        # Defensive parsing: skip malformed tokens rather than failing the notebook
        parts = token.rsplit("-", 1)
        if len(parts) != 2:
            continue
        nid, lab = parts
        if lab not in {"0", "1"}:
            continue
        out.append((nid, int(lab)))
    return out

beh["impressions_list"] = beh["impressions"].apply(parse_impressions)
beh["impressions_len"] = beh["impressions_list"].apply(len)

print("Bad timestamps (NaT):", beh["time"].isna().sum())
beh[["impression_id","user_id","time","history_len","impressions_len"]].head()
```

## Testing

These tests validate:

- impression parsing handles empty strings and malformed tokens
- history parsing yields correct lengths

```python
# Test parse_impressions
assert parse_impressions("") == []
assert parse_impressions(np.nan) == []
assert parse_impressions("N1-1") == [("N1", 1)]
assert parse_impressions("N1-1 N2-0") == [("N1", 1), ("N2", 0)]
assert parse_impressions("BROKEN_TOKEN") == []
assert parse_impressions("N1-x") == []

# Test history parsing
tmp = pd.DataFrame({"history": ["N1 N2 N3", "", None]})
tmp["history"] = tmp["history"].fillna("").astype(str).str.strip()
tmp["history_list"] = tmp["history"].apply(lambda s: s.split() if s else [])
tmp["history_len"] = tmp["history_list"].apply(len)
assert tmp["history_len"].tolist() == [3, 0, 0]

print("Tests passed.")
```

## Distribution of click-history lengths

Goal: understand how much user history is available and how severe cold-start is.

Decision: use 50 bins to balance resolution vs. readability for a long-tailed count distribution.

```python
plt.figure()
plt.hist(beh["history_len"], bins=50)
plt.title("Distribution of User Click History Length")
plt.xlabel("Number of clicked articles in history")
plt.ylabel("Number of impressions (rows)")
plt.show()

beh["history_len"].describe()
```

## Article Popularity

We flatten all clicked news_id from history_list and compute frequency.

Why log-log plot: popularity typically follows a power-law-like distribution; log scales reveal the long tail clearly

```python
all_clicks = pd.Series([nid for lst in beh["history_list"] for nid in lst], dtype="object")
popularity = all_clicks.value_counts()

print("Top 10 most clicked articles (by history clicks):")
display(popularity.head(10).to_frame("clicks"))

plt.figure()
plt.plot(range(1, len(popularity) + 1), popularity.values)
plt.xscale("log")
plt.yscale("log")
plt.title("Article Popularity Distribution (Log-Log)")
plt.xlabel("Rank")
plt.ylabel("Click count")
plt.show()
```

## User-article sparsity

I estimate sparsity as:

[ 1 - \frac{#\text{observed clicks}}{#\text{users} \times #\text{articles}} ]

This matters because sparsity drives feasibility/choice of algorithms (e.g., matrix factorization, GNNs) and impacts bias (popular items dominate learning).

```python
num_users = beh["user_id"].nunique()
num_articles = news["news_id"].nunique()
num_interactions = len(all_clicks)

possible_interactions = num_users * num_articles
sparsity = 1 - (num_interactions / possible_interactions)

print(f"Users: {num_users:,}")
print(f"Articles: {num_articles:,}")
print(f"Clicks (from history): {num_interactions:,}")
print(f"Potential user-article pairs: {possible_interactions:,}")
print(f"Estimated sparsity: {sparsity:.6f}")
```

## Category Distribution


I joined click counts to news.tsv metadata to see where attention concentrates.

I made the decision to group by category (not subcategory) for checkpoint-level granularity.

```python
pop_df = popularity.rename_axis("news_id").reset_index(name="clicks")
merged = pop_df.merge(news[["news_id", "category"]], on="news_id", how="left")

# Decision: treat missing category as its own bucket to quantify metadata coverage
merged["category"] = merged["category"].fillna("UNKNOWN")

category_clicks = merged.groupby("category")["clicks"].sum().sort_values(ascending=False)

display(category_clicks.head(15).to_frame("total_clicks"))

plt.figure()
category_clicks.head(10).iloc[::-1].plot(kind="barh")
plt.title("Top News Categories by Click Volume (history clicks)")
plt.xlabel("Total clicks")
plt.show()
```

# E. Initial Insights and Direction

**Observations and Insights**

User Click History Distribution

- Observation: Most users have short click histories, with a small number of highly active users.
- Insight: The data is highly sparse and engagement is uneven.
- Implication: Models must handle cold-start users and sparse interaction patterns.

Article Popularity

- Observation: A small number of articles receive most of the clicks (long-tail distribution).
- Insight: Strong popularity bias exists.
- Implication: Frequent pattern mining may overemphasize trending articles; debiasing may be necessary.

Interaction Matrix Sparsity

- Observation: The user–article matrix is extremely sparse.
- Insight: Most possible interactions are unobserved.
- Implication: Matrix factorization or graph-based methods are more appropriate than simple co-occurrence counts.

Category Imbalance

- Observation: Certain news categories dominate click volume.
- Insight: User interest is concentrated in specific topics.
- Implication: Topic-aware clustering or recommendation may improve personalization.

Temporal Behavior

- Observation: User clicks occur in bursts with long gaps between sessions.
- Insight: News consumption is session-based and time-sensitive.
- Implication: Sequential or time-aware models may outperform static methods.

Potential Research Questions:

- How does popularity bias affect recommendation diversity?

- Do graph-based embeddings perform better under high sparsity?

- Can temporal modeling improve prediction accuracy?

- Does topic-aware clustering enhance recommendation relevance?
